```python
import numpy as np
import pandas as pd
```

```python
df_canada = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                         sheet_name='Canada by Citizenship',
                         skiprows = range(20),
                         skipfooter = 2)
```

```python
df_canada.head(2)
```

```python
df_canada.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace = True)
```

```python
df_canada.head()
```

```python
df_canada.rename(columns = {'OdName':'Country', 'AreaName': 'Continent','RegName':'Region'}, inplace = True)
```

```python
df_canada.head()
```

```python
df_canada.columns = list(map(str, df_canada.columns))
```

```python
#check columns labebls type
all(isinstance(column, str) for column in df_canada.columns)
```

```python
df_canada.set_index('Country', inplace = True)
```

```python
df_canada.head()
```

```python
df_canada['Total'] = df_canada.sum(axis=1)
```

```python
df_canada.columns.values
```

```python
df_canada.head()
```

```python
data_dimension = df_canada.shape
print(data_dimension)
```

```python
#lets create a list of years
years = list(map(str, range(1980, 2013)))
print(years
     )
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_canada.sort_values(['Total'], ascending = False, axis = 0, inplace= True)

df_top5 = df_canada.head()
#transpose
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area',
             stacked = False , figsize = (20, 10))

plt.title('Immigration trend of top countries')
plt.ylabel('Number of immigrant')
plt.xlabel('Years')
plt.show()
```

```python
# option 1: scripting layer
df_top5.plot(kind='area',
           alpha=0.25,# 0-1, default value a = 0.5,
           stacked=False,
           figsize=(20,10))

plt.title('Immigration trend of top countries')
plt.ylabel('Number of immigrant')
plt.xlabel('Years')
plt.show()
```

```python
#option 2: Artist layer

ax = df_top5.plot(kind='area', alpha=0.35, stacked=False, figsize=(20,10))

ax.set_title('Immigration trend of top countries')
ax.set_ylabel('Number of immigrant')
ax.set_xlabel('Years')
```

```python
#what is the immigration distribution for denmark, norway and sweden for years 1980-2013
df_canada.loc[['Denmark', 'Norway', 'Sweden'],years]
```

```python
df_canada.loc[['Denmark', 'Norway', 'Sweden'],years].plot.hist()
```

```python
df_t = df_canada.loc[['Denmark', 'Norway', 'Sweden'],years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist', figsize=(10,6))

plt.title('Histogram of immigration from denmark, norway and sweden from 1980-2013')
plt.ylabel('Number of years')
plt.xlabel('Number of immigrants')
plt.show()
```

```python
# - increase the bin size to 15 by passing in [bins] parameter
# - set transparency to 60% by passing in alpha parameter
# - label the x-axis by passing in x-label parameter
# - change the colors of the plots by passing in color parameter
count, bin_edges = np.histogram(df_t, 15)

#un-stacked histogram

df_t.plot(kind = 'hist',
         figsize=(10,6),
         bins = 15,
         alpha = 0.6,
         xticks = bin_edges,
         color=['coral', 'darkslateblue', 'mediumseagreen'])
plt.title('Immigration from five countries from 1980-2013 in canada')
plt.ylabel('Number of years')
plt.xlabel('Number of immigrants')
plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)

xmin = bin_edges[0]-10 # first bin value is 31.0, adding buffer of 10 for aesthethic purpose
xmax = bin_edges[-1] +10 # last bin values is 308.0, adding buffer of 10 for aesthetic purpose

#stacked histogram
df_t.plot(kind = 'hist',
         figsize=(10,6),
         bins = 15,
         color=['coral',['darkslateblue'], 'mediumseagreen'],
         stacked = True,
         xlim = (xmin,xmax))
plt.title('Immigration from five countries from 1980-2013 in canada')
plt.ylabel('Number of years')
plt.xlabel('Number of immigrants')
plt.show()
```

```python
# kind = bar -- creates a vertical bar plot
# kind = barh -- create a horizontal bar plot
 
#lets compare the number of Icelandic immigrants () to canada from year 1980-2013

#step 1 :lets get the data

df_ice = df_canada.loc['Iceland', years]
```

```python
df_ice.head()
```

```python
#step : 2 plot data

df_ice.plot(kind = 'bar', figsize=(10,6), color=['Purple'])

plt.title("Iceland immigration")
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_ice.plot(kind = 'bar', figsize=(10, 6), rot=90)# rotate the bars by 90 degree
plt.xlabel('Years')

plt.ylabel('Number of immigrants')
plt.title('Iceland immigration')

#annotate arrow
plt.annotate(' ',
            xy = (32, 72),
            xytext = (28, 20),
            xycoords = 'data',
            arrowprops = dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )

plt.show()
```

```python
df_ice.plot(kind='bar', figsize=(10, 6), rot=90) # rotate the bars by 90 degrees

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

# Annotate arrow
plt.annotate('',                      # s: str. Will leave it blank for no text
             xy=(32, 70),             # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),         # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',         # will use the coordinate system of the object being annotated 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )
# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis', # text to display
             xy=(28, 30),                    # start the text at at point (year 2008 , pop 30)
             rotation=72.5,                  # based on trial and error to match the arrow
             va='bottom',                    # want the text to be vertically 'bottom' aligned
             ha='left',                      # want the text to be horizontally 'left' algned.
            )
plt.show()
```

```python
df_canada.sort_values(by='Total', ascending =True, inplace= True)
df_top15 = df_canada['Total'].tail(15)
```

```python
df_top15.head()
```

```python
df_top15.plot(kind='barh', figsize=(12,12), color=['steelblue'])
plt.xlabel('Number of immigrants')
plt.title('Top 15 countries')

#annotate values labels to each country

for index, value in enumerate(df_top15):
    label = format(int(value), ',') #format with commas
    
    #place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
    plt.annotate(label, xy=(value - 47000, index - .10), color = 'white')

plt.show()
```
